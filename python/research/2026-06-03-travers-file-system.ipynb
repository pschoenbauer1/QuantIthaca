```python

def get_dir_size(path):
    sizes_ = {}

    def dir_size(path):
        total = 0
        try:
            with os.scandir(path) as it:
                for entry in it:
                    try:
                        if entry.is_dir(follow_symlinks=False):
                            total += dir_size(entry.path)
                        else:
                            total += entry.stat(follow_symlinks=False).st_size
                    except (PermissionError, FileNotFoundError):
                        pass
        except (PermissionError, FileNotFoundError):
            pass

        sizes_[path] = total
        return total

    dir_size(path)
    
    return sizes_


sizes2 = get_dir_size("C:\\Users\\phili\\OneDrive\\.fotos")

for path, size in sorted(sizes2.items(), key=lambda x: x[1], reverse=True)[:100]:
    print(f"{size/1024**3:10.2f} GB  {path}")
```

```python
import pandas as pd

# path columns
df = data[data["size"] > 1]

path_cols = [0, 1, 2, 3, 4, 5, 6, 7]

# Build full path
df["path"] = (
    df[path_cols]
    .fillna("")
    .astype(str)
    .apply(lambda r: "\\".join(x for x in r if x), axis=1)
)

# Lookup size by path
size_lookup = dict(zip(df["path"], df["size"]))

# Build tree
tree = {}

for path in df["path"]:
    parts = [p for p in path.split("\\") if p]
    node = tree
    for part in parts:
        node = node.setdefault(part, {})

# Collect rows in size-descending DFS order
rows = []

def walk(node, prefix_parts):
    children = []

    for name, subtree in node.items():
        parts = prefix_parts + [name]
        full_path = "\\".join(parts)
        size = size_lookup.get(full_path, 0)

        children.append((name, subtree, parts, size))

    # largest first
    children.sort(key=lambda x: x[3], reverse=True)

    for name, subtree, parts, size in children:
        rows.append(parts + [size])
        walk(subtree, parts)

walk(tree, [])

# Determine max depth
max_depth = max(len(r) - 1 for r in rows)

# Create dataframe
result = pd.DataFrame(
    [
        r[:-1] + [None] * (max_depth - len(r) + 1) + [r[-1]]
        for r in rows
    ],
    columns=[f"level_{i}" for i in range(max_depth)] + ["size"]
)

pd.options.display.max_rows = 200
result.fillna("")
```
